```python
# Dependencies and Setup
import pandas as pd


# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data=purchase_data.rename(columns={"SN":"Player"})
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
total_players=purchase_data["Player"].nunique()
players_number_df=pd.DataFrame({"Total Players":[total_players]})
players_number_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_item_number=purchase_data["Item Name"].nunique()
average_price='${:,.2f}'.format(purchase_data["Price"].mean())
purchase_number=purchase_data["Purchase ID"].count()
total_revenue='${:,.2f}'.format(purchase_data["Price"].sum())

summary_data=pd.DataFrame([{
    "Number of Unique Items":unique_item_number,
    "Average Price":average_price,
    "Number of Purchases":purchase_number,
    "Total Revenue":total_revenue
    }])
summary_data
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# purchase_data["Gender"].value_counts()
unique_players=purchase_data.groupby("Gender")["Player"].nunique().sort_values(ascending=False)
unique_players
```

```python
total_gender=unique_players.sum()

# percent_male=(unique_players[0]/total_gender).map('{:,.2%}'.format)
# percent_female=(unique_players[1]/total_gender).map('{:,.2%}'.format)
# percent_other=(unique_players[2]/total_gender).map('{:,.2%}'.format)
# percent_players=[percent_male,percent_female,percent_other]


percent_male='{:,.2%}'.format(unique_players[0]/total_gender)
percent_female='{:,.2%}'.format(unique_players[1]/total_gender)
percent_other='{:,.2%}'.format(unique_players[2]/total_gender)
percent_players=[percent_male,percent_female,percent_other]

gender_demographics_df=pd.DataFrame({"Total Count":unique_players,
    "Percentage of Players":percent_players},
    index=["Male","Female","Other / Non-Disclosed"])
gender_demographics_df

```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_count=purchase_data["Gender"].value_counts().sort_index()
average_purchase_price=purchase_data.groupby("Gender")["Price"].mean().map('${:,.2f}'.format)
total_purchase_value=purchase_data.groupby("Gender")["Price"].sum()
purchase_df=pd.DataFrame({"Purchase Count":purchase_count,"Average Purchase Value":average_purchase_price,
"Total Purchase Value":total_purchase_value})
average_purchase_per_person=(purchase_df["Total Purchase Value"]/gender_demographics_df["Total Count"]).map('${:,.2f}'.format)
purchase_df["Total Purchase Value"]=total_purchase_value.map("${:,.2f}".format) # To convert from float to currency
purchase_df["Avg Total Purchase per Person"]=average_purchase_per_person
purchase_df
```

## Drop Duplicates

```python
duplicateRowsDF = purchase_data[purchase_data.duplicated(['Player'])]
unique_players=purchase_data.drop_duplicates(['Player'])
unique_players_df= pd.DataFrame(unique_players)
unique_players_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins=[0,9,14,19,24,29,34,39,100]
age_labels=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
# print(purchase_data["Age"])
age_range=pd.cut(unique_players_df["Age"],bins,labels=age_labels)
age_demographics_df=pd.DataFrame({"Total Count":age_range.value_counts().sort_index()})
percentage_of_players=(age_demographics_df["Total Count"]/(age_demographics_df["Total Count"].sum())).map('{:,.2%}'.format)
# percentage_of_players
age_demographics_df["Percentage of Players"]=percentage_of_players # Adding the column of "Percentage of Players"
age_demographics_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python

age_range_non_unique=pd.cut(purchase_data["Age"],bins,labels=age_labels)
purchase_count_per_age=age_range_non_unique.value_counts().sort_index()

total_purchase_per_age=purchase_data.groupby([age_range_non_unique])["Price"].sum()

average_purchase_price_per_person=total_purchase_per_age/purchase_count_per_age

average_tot_purchase_per_person=total_purchase_per_age/age_demographics_df["Total Count"]

purchasing_analysis_df=pd.DataFrame({"Age Ranges":age_labels,
    "Purchase Count":purchase_count_per_age,
    "Average Purchase Price":average_purchase_price_per_person.map('${:,.2f}'.format),
    "Total Purchase Value":total_purchase_per_age.map('${:,.2f}'.format),
    "Avg Total Purchase per Person":average_tot_purchase_per_person.map('${:,.2f}'.format)})

purchasing_analysis_df
```
